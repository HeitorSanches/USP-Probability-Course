## Random!

```python

coin = ["head","tails"]
for i in range(0,10):
    # flip coin
    output = random.choice(coin)
    print(output)
```

```python
coin = ["head","tails"]
lheads = []
ln = []

for n in range(1,1000):
    nheads = 0
    for i in range(0,n):
        side = np.random.choice(coin)
        if side == "head":
            nheads += 1
    fh = nheads/n
    lheads.append(fh)
    ln.append(n)

plt.plot(ln,lheads)
plt.axhline(y=0.5,color="r",linestyle = '-')
plt.xlabel("Number of flips")
plt.ylabel("Frequency of Heads")
plt.show()

```

```python
dice =[1,2,3,4,5,6]

n = 1000
vfaces = np.zeros(6)

for i in range(0,n):
    face = np.random.choice(dice)
    vfaces[face-1] += 1

print(f"theoric value : {1/6}")
print(f"frequencies obtained : {vfaces/n}") 
```

### Unfair coin

```python
coin = ["head","head","tails","tails","tails","tails","tails","tails","tails","tails"]
p = 0
for c in coin:
    if c == "head":
        p += 1
p = p /len(coin)
lheads = []
ln = []

for n in range(1,1000):
    nheads = 0
    for i in range(0,n):
        side = np.random.choice(coin)
        if side == "head":
            nheads += 1
    fh = nheads/n
    lheads.append(fh)
    ln.append(n)

plt.plot(ln,lheads)
plt.axhline(y=p,color="r",linestyle = '-')
plt.xlabel("Number of flips")
plt.ylabel("Frequency of Heads")
plt.show()

```

## Frequentist Probability

```python
cf= 1000 # count of flips
coin = ['head','tails']
nh = 0 #number of heads

for f in range(0,cf):
    output = random.choice(coin)
    if(output=='head'):
        nh +=1

print(f"Probability : {nh/cf}")
```

```python
urn = ["W"] * 3000 + ["B"] * 2000 # W = White ball , #B = Black Ball
# We will consider 3 thousand white balls and 2 thousand black balls
nsim = 100 # number of simulations
fB = 0 #frequency of black balls

for i in range(0,nsim):
    ball = random.choice(urn)
    if ball == "B":
        fB+=1

fB = fB/nsim

print(f"Frequency of black balls : {fB}")
```
