## Random!

```python
import random
coin = ["head","tails"]
for i in range(0,10):
    # flip coin
    output = random.choice(coin)
    print(output)
```

## Frequentist Probability

```python
cf= 1000 # count of flips
coin = ['head','tails']
nh = 0 #number of heads

for f in range(0,cf):
    output = random.choice(coin)
    if(output=='head'):
        nh +=1

print(f"Probability : {nh/cf}")
```
